```python
import pandas as pd
import numpy as np
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header=None)
df
```

```python
import pandas as pd

# Define column names from UCI dataset description
columns = [
    "symboling", "normalized_losses", "make", "fuel_type", "aspiration",
    "num_doors", "body_style", "drive_wheels", "engine_location",
    "wheel_base", "length", "width", "height", "curb_weight",
    "engine_type", "num_cylinders", "engine_size", "fuel_system",
    "bore", "stroke", "compression_ratio", "horsepower",
    "peak_rpm", "city_mpg", "highway_mpg", "price"
]

# Read the data
df = pd.read_csv(url, names=columns)

# Check first rows
df
```

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['normalized_losses']
```

```python
avg_norm_loss = df['normalized_losses'].astype('float').mean(axis=0)

avg_norm_loss
```

```python
df.dtypes
```

```python
df["normalized_losses"].replace(np.nan, avg_norm_loss, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['num_doors']
```

```python
df['num_doors'].value_counts()
```

```python
df['num_doors']= df['num_doors'].fillna('four')
df['num_doors']
```

```python
df.isnull().sum()
```

```python
df['bore']
```

```python
avg_bore = df['bore'].astype('float').mean(axis = 0).round(2)
avg_bore
```

```python
df['bore'].replace(np.nan, avg_bore, inplace = True)
df['bore']
```

```python
df.isnull().sum()
```

```python
df['stroke']
```

```python
avg_str = df['stroke'].astype('float').mean(axis = 0).round(2)
avg_str
```

```python
df['stroke'].replace(np.nan, avg_str, inplace = True)
df['stroke']
```

```python
df.isnull().sum()
```

```python
df['horsepower']
```

```python
avg_hrp = df['horsepower'].astype('float').mean(axis = 0).round(2)
avg_hrp
```

```python
df['horsepower'].replace(np.nan, avg_hrp, inplace = True)
df['horsepower']
```

```python
df.isnull().sum()
```

```python
df['peak_rpm']
```

```python
avg_rpm = df['peak_rpm'].astype('float').mean(axis = 0).round(2)
avg_rpm
```

```python
df['peak_rpm'].replace(np.nan, avg_rpm, inplace = True)
df.isnull().sum()
```

```python
df.dropna(subset=['price'], inplace = True)
df.isnull().sum()
```

```python
df.head()
```

### Question #1:
According to the example above, transform mpg to L/100km in the column of "city-mpg" and change the name of column to "city-L/100km".

```python
df.dtypes
```

```python
df.rename(columns={'city_mpg': 'city_L_100km'}, inplace=True)
df['city_L_100km'] = 235 / df['city_L_100km']
df.head()
```

## Question #2:
According to the example above, transform mpg to L/100km in the column of "highway-mpg" and change the name of column to "highway-L/100km".

```python
df.rename(columns={'highway_mpg': 'highway_L_100km'}, inplace=True)
df['highway_L_100km'] = 235 / df['highway_L_100km']
df.head()
```

```python
df[['length', 'width', 'height']]
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 
df[['length', 'width', 'height']]
```

```python
df.dtypes
```

```python
df["horsepower"] = df["horsepower"].astype('int')
df["horsepower"]
```

```python
bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]), 4 )
bins
```

```python
group = ["Low", "Medium","High"]
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel_type"])
dummy_variable_1.head()
```

```python
# merge data frame "cars" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel_type", axis = 1, inplace=True)
```
